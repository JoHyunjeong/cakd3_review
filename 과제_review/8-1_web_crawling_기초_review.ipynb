```python
# 도전 과제
# Q. 네이버 영화 랭킹 가져와서 첫번째 영화제목을 출력하세요.
import re
import urllib

URL ='https://movie.naver.com/movie/sdb/rank/rmovie.nhn'
response = urllib.request.urlopen(URL)
response = response.read().decode('utf-8')

response = response.split('<a href="/movie/bi/mi/basic.naver?code=')[1]
response = re.search(r'>.*<',response)

print(response.group().strip('><'))
```

```python
# 도전 과제
# Q. 네이버 영화 랭킹 가져와서 전체 영화제목을 출력하세요.
import re
import urllib

URL ='https://movie.naver.com/movie/sdb/rank/rmovie.nhn'
response = urllib.request.urlopen(URL)
response = response.read().decode('utf-8')

for i in range(1,51):
    result = response.split('<a href="/movie/bi/mi/basic.naver?code=')[i]
    movie = re.search(r'>.*<',result)
    print(f"{i}위 : {movie.group().strip('><')}")
```
