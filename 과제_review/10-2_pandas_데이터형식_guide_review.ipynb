Q. 5행 5열의 리스트, 배열, 사전, 형태의 데이터를 데이터프레임으로 변환하여 출력하세요.

```python
import pandas as pd

# 리스트 -> 데이터프레임

data1 = [[38,13,73,10,76],[6,80,65,17,2],[77,72,7,26,51],[221,19,85,12,29],[30,15,51,69,88]]
df1 = pd.DataFrame(data1,index = ['학생1','학생2','학생3','학생4','학생5'],
               columns = ["현실형","관습형","진취형","사회형","탐구형"])
print(type(df1))
df1
```

```python
import pandas as pd
import numpy as np

# 배열 -> 데이터프레임
np.random.seed(1)
data2 = np.random.randint(1,101,size = (5,5))
df2 = pd.DataFrame(data2,index = ['학생1','학생2','학생3','학생4','학생5'],
               columns = ["현실형","관습형","진취형","사회형","탐구형"])
print(type(df2))
df2
```

```python
# 사전 -> 데이터프레임
data3 = {'현실형':[38,6,77,21,30],'관습형':[13,80,72,19,15],'진취형':[73,65,7,85,51],'사회형':[10,17,26,12,69],'탐구형':[76,2,51,29,88]}
df3 = pd.DataFrame(data3,index = ['학생1','학생2','학생3','학생4','학생5'])
print(type(df3))
df3

### dictionary : key는 column명, value는 리스트 또는 배열 형식으로!
```

Q. 위에서 변환한 데이터프레임에서 데이터 변경, 삭제, 전치, 인덱스 등 처리를 한 후 다시 리스트, 배열, 사전 형태로 변환하여 출력하세요.

```python
# 복사
df=df1.copy()
print(type(df))
df
```

```python
# 데이터 변경
df.iloc[3,0]=21
df
```

```python
# 전치
df=df.transpose()
df
```

```python
# 각 열의 최대값(value) - 사용 안 함
# f=lambda x:x.max()
# df['max_value']=df.apply(f,axis=1) 
# df
### 또는 df.max(axis=1)
```

```python
# 각 열의 최대값 행 이름(index) 찾기 + 시리즈로 만들기
max=df.idxmax()
max_value=pd.Series(max)
max_value
```

```python
# 다시 전치(원위치)
df=df.transpose()
df
```

```python
# 열 추가
df = pd.concat([df,max_value],axis=1) 
df
```

```python
# 컬럼이름 변경
df=df.rename(columns={0:'대표흥미코드'},inplace=0)
df
```

```python
# 삭제
df = df.drop('학생5',axis=0)
df
```

```python
# 데이터프레임 -> 배열
ar = df.values
print(ar,type(ar))
```

```python
# 데이터프레임 -> 리스트
li = ar.tolist()
print(li,type(li))
```

```python
# 또 다시 전치(딕셔너리 변환하기 위해)
df=df.transpose()
df
```

```python
# 데이터프레임 -> 사전
dict = df.to_dict()
print(dict,type(dict))
```
