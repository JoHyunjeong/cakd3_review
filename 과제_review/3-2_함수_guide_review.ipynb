Q. 1000명에 대하여 아래와 같이 출력하세요  
bmi = round(int(weight) / ((int(height) / 100) ** 2),2)  
이름 :라바  
몸무게: 80  
키: 146  
BMI: 37.53049352598987  
결과: 과체중  
    
BMI Categories:  
Underweight = <18.5  
Normal weight = 18.5–24.9  
Overweight = 25–29.9  
Obesity = BMI of 30 or greater'  

```python
import random

hanguls = '가나다라마바사아자차카타파하'
with open('info.txt','w') as f:
    for i in range(1, 1001):
        name = random.choice(hanguls) + random.choice(hanguls)
        weight = random.randrange(40,100)
        height = random.randrange(140,200)
        f.write('{},{},{}\n'.format(name, weight, height))
```

```python
with open('info.txt','r') as f:
    lines = f.readlines()
    for line in lines:
        name, weight, height = line.split(',')
        bmi = round(int(weight)/((int(height)/100)**2),2)
        if bmi <= 18.5:
            result = 'Underweight'
        elif 18.5 < bmi <= 24.9:
            result = 'Nomalweight'
        elif 25.0 <= bmi <= 29.9:
            result = 'Overweight'
        else:
            result = 'Obesity'
        print('이름 : {}\n몸무게 : {}\n키 : {}BMI : {}\n결과 : {}\n'.format(
        name, weight, height, bmi, result))
```

Q. 사칙연산 기능을 포함하는 모듈 cal4.py 을 작성 후 임포트해서 사칙연산을수행하세요.

```python
with open('cal4.py','w') as f:
    f.write('''
def cal(choice, *args):
    if choice == '+':
        result = 0
        for i in args:
            result += i
        print(result)
    elif choice == '*':
        result = 1
        for i in args:
            result *= i
        print(result)
    elif choice == '-':
        for n,i in enumerate(args):
            if n == 0:
                result = i
            else:
                result -= i
        print(result)
    elif choice == '/':
        try:
            for n,i in enumerate(args):
                if n == 0:
                    result = i
                else:
                    result /= i
            print(result)
        except:
            print('다시 입력해주세요.')
    else:
        print('다시 입력해주세요.')
''')            
```

만든 후 !python cal4.py  실행

```python
from cal4 import cal

cal('/',10,5,2)
cal('/',10,5,0)
cal('+',10,5,2)
```
