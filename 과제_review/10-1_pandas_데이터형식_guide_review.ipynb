사전으로 데이터프레임 생성(8행 5열)

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'col1':np.array([1,2,3,4,5,6,7,8]),
    'col2':np.array(['A','B','C','D','E','F','G','H']),
    'col3':np.array(['가','나','다','라','마','바','사','아']),
    'col4':np.array(['사과','딸기','포도','망고','복숭아','레몬','귤','체리'],),
    'col5':np.array(['빨','주','노','초','파','남','보','검'])})
    
df
```

Q. 생성된 데이터프레임에서 아래 사항을 수행하세요.
- 열이름과 행이름 변경
- 3행만 출력
- 5열만 출력
- 3행 5열의 값 출력
- 4열의 값을 모두 0으로 변경
- 2,3행의 3열 값을 조회(2가지 방법)
- 6행의 2,3,4열의 값을 조회(2가지 방법)
- 2열을 Series와 DataFrame으로 각각 출력
- 2행 3열의 값을 2행 4열의 값과 동일하게 변경
- 2행 3열의 값과 2행 4열의 값을 동시에 만족하는 3,4열을 출력

```python
# 열이름과 행이름 변경

df.rename(columns = {'col1':'int','col2':'alp','col3':'kor','col4':'fru','col5':'color'},inplace=True)

for i in df.index:
    df.rename(index={i:"{}행".format(i+1)},inplace=True)

df
```

```python
df1 = df.copy()
df1
```

```python
# 3행만 출력
print(df1.loc['3행'])
```

```python
print(df1.iloc[2])
```

```python
# 5열만 출력
print(df1["color"])
```

```python
print(df1.iloc[:,4])
```

```python
# 3행 5열의 값 출력
print(df1.loc["3행","color"])
```

```python
print(df1.iloc[2,4])
```

```python
# 4열의 값을 모두 0으로 변경
df1['fru']=0
df1
```

```python
df1.iloc[:,3] = 0
df1
```

```python
# 2,3행의 3열 값을 조회(2가지 방법)
print(df1.query('index=="2행" | index=="3행"')["kor"])

print('-'*50)

print(df1.iloc[1:3,2])

print('-'*50)

print(df1.loc[["2행","3행"]]["kor"])
```

```python
# 6행의 2,3,4열의 값을 조회(2가지 방법)
print(df1.query("index=='6행'")[["alp","kor","fru"]])

print('-'*50)

print(df1.iloc[5,1:4])

print('-'*50)

print(df1.loc[["6행"]][["alp","kor","fru"]])

### 전 문제에서 4열을 0으로 바꿔서 0 출력됨
```

```python
# 2열을 Series와 DataFrame으로 각각 출력
print(df1.alp, type(df1.alp))

print('-'*80)

print(pd.DataFrame(df1['alp']), type(pd.DataFrame(df1['alp'])))
print(df1[['alp']], type(df1[['alp']]))
```

```python
# 2행 3열의 값을 2행 4열의 값과 동일하게 변경
df1.iloc[1,2] = df1.iloc[1,3]
print(df1.iloc[1,2])

print('-'*50)

print(df1)
```

```python
# 2행 3열의 값과 2행 4열의 값을 동시에 만족하는 3,4열을 출력
df1.query("kor==0 & fru==0")[["kor","fru"]]
```

```python
df1.loc[df1['kor']==df1['fru'],['kor','fru']]
```
