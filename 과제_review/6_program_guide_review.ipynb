```python
# pop()이용
from random import shuffle
from time import sleep

gnum=int(input("로또 회수를 입력하세요> "))
for i in range(gnum):
    lt = [x+1 for x in range(45)] # lt = [x for x in range(1,46)]과 같음.
    ret=[]
    for j in range(6): # for문 속 for문 -> 다중 for문
        shuffle(lt) # 리스트 섞은 후
        num=lt.pop() # 마지막꺼 하나 뽑고 리스트 안에서 삭제(중복 제거)
        ret.append(num) # ret리스트에 추가
    ret.sort() # 순서대로 정리
    print('로또번호 %d:'%(i+1),end=' ')
    print(ret)
    sleep(1)
```

Q11. 주어진 데이터에서 연도별 성별 출생아 수를 계산하여 출력하세요.

```python
total=[]
for y in range(1880,2020):
    with open('data/names/yob%d.txt'%y,'r') as f:
        data=[]
        while 1:
            line=f.readline()
            if not line:break
            splitdata=line.split(',')
            data.append(splitdata)     ### 여기까지는 Q.10과 동일
    sum_m=0                            
    sum_f=0
    for i in range(len(data)):
        if data[i][1] == 'F':         ### data 리스트 속 각 리스트는[이름,성별,출생아 수]로 되어 있으므로
            sum_f+=int(data[i][2])    ### if문을 사용하여 성별을 나타내는 각 리스트의 1번 인덱스가 F인지 아닌지로 나눠서
        else:                        ### F(여자)면 sum_f에, 아니면(나머지=M, 남자) sum_m에 출생아 수인 2번 인덱스 값을 합산함.
            sum_m+=int(data[i][2])
    total.append([y,sum_m,sum_f])    ### 마지막으로 total리스트에 연도,여자 수,남자 수를 리스트로 묶어서 넣음
                                     ### 여기까지 하나의 포문이므로 최종적으로 total리스트에 연도별로 성별 출생아 수가 들어감
for i in range(len(total)):
    print("{}년도 : 남 {} 명, 여 {} 명".format(total[i][0],total[i][1],total[i][2]))
```

```python
for y in range(1880,2020):
    with open('data/names/yob%d.txt'%y,'r') as f:
        count_female = count_male = 0
        while True:
            line = f.readline()
            if not line: break        
            count = line.split(",")
            if count[1] =="M":
                count_male += int(count[2])
            else:
                count_female += int(count[2])

    print("{}년 남자 출생아 수 : {}, 여자 출생아 수 : {}".format(y, count_male, count_female))
```

Q12. 출생년도를 입력하면 그해 인기순 상위10개 이름을 출력하세요.

```python
y=int(input("출생년도 입력>"))
with open('data/names/yob%d.txt'%y,'r') as f:
    data=[]
    while 1:
        line=f.readline()
        if not line:break
        splitdata=line.split(',')
        data.append(splitdata)
    f=[]
    m=[]
    for i in range(len(data)):
        if data[i][1] == 'F':
            f.append([data[i][2],data[i][0]])            
        else:
            m.append([data[i][2],data[i][0]])
print("\n<{}년도 인기있는 이름 TOP 10>\n".format(y))
for i in range(0,10):
    print("\n{}위 - 여자 : {}, 남자 : {}".format(i+1,f[i][1],m[i][1]))
    
### 파일이 출생자 수 내림차순으로 정리되어 있길래 순서대로 10개씩 출력함😅
```

```python
from os.path import exists

def getTop10BabyName(year):
    nameF={}
    nameM={}
    
    filename='data/names/yob%s.txt'%year
    if not exists(filename):
        print('[%s] 파일이 존재하지 않습니다.'%filename)
        return None
    with open(filename,'r') as f:
        data = f.readlines()
        for d in data:
            if d[-1] == '\n':   ### 끝에 개행문자 있으면 처리!
                d=d[:-1]
                
            tmp = d.split(',') ### 콤마를 기준으로 나눠서 리스트에 저장
            name = tmp[0]
            sex = tmp[1]
            birth = tmp[2]
            
            if sex=='F':       ### 딕셔너리에 이름을 key, 출생아 수를 value로 저장
                ret=nameF
            else:
                ret=nameM
                
            if name in ret:    ### 이름이 중복인 경우를 위해서 추가
                ret[name]+=int(birth)  
            else:
                ret[name]=int(birth)
    retF = sorted(nameF.items(),key=lambda x:x[1],reverse=True)  ### items()->튜플로 key와 value 줌. 여기서 1번 인덱스는 value
    retM = sorted(nameM.items(),key=lambda x:x[1],reverse=True)
    
    for i,name in enumerate(retF):
        if i > 9: 
            break
        print('TOP_%d 여자아기 이름:%s'%(i+1,name))
    for i,name in enumerate(retM):
        if i > 9: 
            break
        print('TOP_%d 남자아기 이름:%s'%(i+1,name))
y=input('인기순 상위 10개 이름을 알고 싶은 출생년도를 입력하세요> ')
getTop10BabyName(y)
```
