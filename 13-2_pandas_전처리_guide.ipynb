```python
# Q. age 컬럼을 범주화하여 'age_class' 컬럼을 생성하세요.
# 범주화

# Null값은 별도처리 필요

def agegroup(x):
    cat=''
    if x < 10:
        cat = 'child'
    elif x < 25:
        cat = 'young'
    elif x < 40:
        cat = 'adult'
    elif x < 60:
        cat = 'senior'
    else:
        cat = 'old'
    return cat

age_class = tdf.age.apply(agegroup)
tdf['age_class'] = age_class
tdf
```

```python
import pandas as pd
fifa = pd.read_csv('FIFA.csv',encoding='cp949')
fifa
```

위 표의 Height는 피트 (1피트 = 30 cm)

```python
# Height 값을 cm로 변환하여 새로운 컬럼 Height_cm을 생성하세요.
def ht_ex(x):
    f, i = map(int, x.split('\''))
    return f*30.48 + i*2.54

fifa['Height_cm'] = fifa['Height'].apply(ht_ex)
fifa.head()
```

```python
fifa = fifa.drop('Height',axis = 1)
fifa.head()
```
