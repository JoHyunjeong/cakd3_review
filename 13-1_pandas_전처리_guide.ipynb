```python
import numpy as np
import pandas as pd
data1 = np.arange(1,21).reshape(4,5)
data2 = np.arange(11,31).reshape(5,4)

df1 = pd.DataFrame(data1, columns=list('abcde'))
df2 = pd.DataFrame(data2, columns=list('dehi'))

print(df1,'\n')
print(df2)
```

```python
### 행 방향으로 합치기
con1 = pd.concat([df1,df2],ignore_index=True)
con1

### 컬럼명이 겹치는 곳은 쭉 합쳐지고 컬럼명 겹치지 않는 부분은 NaN값으로 채우면서 합침!
# NaN : Not a Number 
```

```python
con2 = pd.concat([df1,df2],axis=1)
con2
```

```python
df1['ind'] = [1,2,3,4]
df1
```

```python
df2
```

```python
df2['ind'] = [1,2,3,4,5]
df2
```

```python
# merge()를 이용한 결합
pd.merge(df1,df2,on='ind', how='outer') ### 공통된건 on으로 씀(기준), how는 어떻게 할지(outer는 합집합, inner는 교집합) 
```

```python
# merge()를 이용한 결합
pd.merge(df1,df2,on='ind') # how=inner가 default
```

```python
# merge()를 이용한 결합
pd.merge(df1,df2,on='ind', how='inner')
```

```python
df1 = df1.drop('ind',axis=1)
df2 = df2.drop('ind',axis=1)
```

```python
# join은 인덱스 기준으로 결합(default는 교집합)
df1.join(df2,lsuffix='_1', rsuffix='_2')  # 겹치는 컬럼명에 구분할 이름 붙여주기 / lsuffix와 rsuffix 의 l과 r은 왼쪽 오른쪽
```

```python
df1.join(df2,lsuffix='_1', rsuffix='_2', how='inner') ### default가 inner 이므로 쓰나 안쓰나 동일!
```

```python
df1.join(df2,lsuffix='_1', rsuffix='_2', how='outer') ### outer를 주면 합집합이기 때문에 공통아닌 곳은 NaN값으로 나옴
```

```python
df1['ind'] = [1,2,3,4]
df2['ind'] = [1,2,3,4,5]
```

```python
df1.set_index('ind').join(df2.set_index('ind'),
                          lsuffix='_1', rsuffix='_2', on='ind',how='outer') # on을 주면 컬럼으로 나옴
```
