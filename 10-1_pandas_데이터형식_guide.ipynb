### 판다스

- 판다스는 구조화된 데이터 형식을 제공. 시리즈는 1차원 배열, 데이터프레임은 2차원 배열
- 라이브러리는 여러 종류의 class와 다양한 내장 함수로 구성. 시리즈와 데이터프레임은 대표적인 클래스 객체임
- 시리즈 인덱스는 데이터 값과 일대일 대응. 파이썬 딕셔너리와 비슷한 구조. pandas.Series(딕셔너리)
- Contents
    - Series(변환, 인덱스 구조, 원소 선택)
    - DataFrame(변환, 행인덱스/열이름 지정, 삭제, 선택(###특정조건), 추가, 변경, 전치(###행과 열 바꿈), 인덱스 활용)

- 시리즈와 데이터프레임은 판다스에 들어있는 class  
- 판다스는 데이터 전처리할 때 중요한 tool

#### 데이터프레임
- 데이터프레임은 2차원 배열. R의 데이터프레임에서 유래.
- 데이터프레임의 열은 각각 시리즈 객체.
- 시리즈를 열벡터라고 하면 데이터프레임은 여러개의 열벡터들이 같은 행 인덱스를 기준으로 줄지어 결합된 2차원 벡터 또는 행렬.
- 선형대수학에서 열 벡터(m x 1 행렬)는 m 원소들의 단일 열 행렬
- 행 벡터(1 x m 행렬)은 m원소들의 단일 행 행렬.
- 리스트, 딕셔너리, ndarray 등 다양한 데이터로부터 생성
- 반대로 리스트, 딕셔너리, ndarray 등으로 변환될 수 있음

```python
# 배열을 데이터프레임으로 변환
np.random.seed(0) # 랜덤 값 고정
data = np.random.randint(100,120,size=(3,3))
print(data, type(data))

df = pd.DataFrame(data,index=['d1','d2','d3'],
                 columns = ['pd','sales','inv'])

df
```

```python
# pandas indexing : iloc 정수 인덱스, loc 이름 인덱스
print(df.iloc[1]) # 시리즈로 출력됨
print('-'*50)
print(df.loc['d2'])
print('-'*50)
print(df.iloc[1].loc['pd'])
print('-'*50)
print(df.iloc[1,0])
print('-'*50)
print(df.loc['d2','pd'])
print('-'*50)
print(df.pd)
```

```python
import numpy as np
import pandas as pd

a = np.random.randint(1,5,size=(10,5))
print(a,type(a))
```

```python
# 2차원 리스트를 데이터프레임으로 변환
list1 = a.tolist()
print(list1,type(list1))
df1 = pd.DataFrame(list1, columns=['c1','c2','c3','c4','c5'])
df1.head(3)
```

```python
import pandas as pd
import numpy as np

df_1 = pd.DataFrame({
    'col1':np.array([1,2,3]),
    'col2':np.array(['A','B','C'])
})
df_2 = pd.DataFrame({
    'col1':np.array([4,5,6]),
    'col2':np.array(['D','E','F'])
})
print(df_1)
print(df_2)
```

```python
# Q. df_1과 df_2를 행방향과 열방향으로 병합하여 출력하세요.

df_3 = pd.concat([df_1,df_2],axis=1) # 열방향
df_4 = pd.concat([df_1,df_2]) # 행방향(default)
print(df_3)
print(df_4)
```

```python
# 데이터 추출
print(df_4.col2, type(df_4.col2)) # 컬럼 하나만 뽑으면 Series
print('-'*50)

print(df_4['col2'], type(df_4['col2']))
```

```python
# 컬럼 하나만 뽑을 때 Series가 아닌 DataFrame으로 출력
df = pd.DataFrame(df_4['col2'])
print(df,type(df))
print('-'*50)

print(df_4[['col2']], type(df_4[['col2']]))
```
