셀레니움 함수
- find_element_by_id : id 속성을 사용하여 접근
- find_element(s)_by_class_name : 클래스를 사용하여 접근
- find_element(s)_by_name : name 속성을 사용하여 접근
- find_element(s)_by_xpath : xpath 속성을 사용하여 접근
- find_element(s)_by_link_text : 앵커대크(a 태그)에 사용되는 텍스트로 접근
- find_element(s)_by_partial_link_text :  앵커태그(a 태그)에 사용되는 일부 텍스트로 접근
- find_element(s)_by_tag_name : 태그를 사용하여 접근
- find_element(s)_by_css_selector : CSS 선택자를 사용하여 접근

(s)는 하나 찾을 때 필요 없고 여러개 찾을 때 붙임! 붙이면 for문 써서 출력

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time


url = "https://google.com"
driver = webdriver.Chrome("/Users/hyunjeongjo/Desktop/chrome/chromedriver") ## 객체 만들기
driver.get(url)

search_box = driver.find_element_by_xpath("/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input")
search_box.send_keys('빅데이터')
search_box.send_keys(Keys.RETURN)

elements = driver.find_elements_by_xpath('//*[@id="rso"]/div/div/div/div/a/h3')

for e in elements:
    print(e.text)
    print(e.text, file=open('test_1.txt', 'a', encoding='utf-8'))
time.sleep(3)
driver.close()
```

```python
# requests + bs4

import requests
from bs4 import BeautifulSoup

url = "https://naver.com"
response = requests.get(url)
html = response.text

bs = BeautifulSoup(html,"lxml") ## html.parser 또는 lxml

titles = bs.select('#themecast > div.theme_cont > div > div > ul > li > a.theme_info > strong')
for i,t in enumerate(titles):
    if i < 5:
        print(t.text)
    else:
        pass
```
