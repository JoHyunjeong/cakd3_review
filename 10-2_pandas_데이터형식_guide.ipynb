Q. 딕셔너리로 5행 5열 데이터프레임을 작성하세요.

```python
import pandas as pd
#딕셔너리를 정의
dict_data = {'a':[1,2,3,4,5],'b':[6,7,8,9,10],'c':[11,12,13,14,15],'d':[16,17,18,18,20],'e':[21,22,23,24,25]}

# 딕셔너리를 데이터프레임으로 변환. 인덱스를[r0, r1, r2]로 지정
df = pd.DataFrame(dict_data, index=['r0','r1','r2','r3','r4'])
df
```

```python
# Q. df에 r5,r6 2개의 행을 추가하고 값은 0을 적용하여 출력하세요.
df.loc['r5']=0
df.loc['r6']=0  ### loc 없이 바로 df.['x']=y 로 하면 열 추가됨!
df
```

```python
# Q. df에 r7, r8 2개의 행을 다른 방법으로 추가
new_index = ['r0','r1','r2','r3','r4','r5','r6','r7','r8']
df5 = df.reindex(new_index,fill_value=0) ### fill_value=x 는 새로 추가되는 것만 해당
df5
```

```python
# Q. r5 ~ r8까지 삭제 후 출력하세요.
df1 = df5.drop(['r5','r6','r7','r8'],axis=0) ### 행 삭제
df1
```

```python
### 원본수정하려면 
df5.drop(['r5','r6','r7','r8'],axis=0, inplace=True) 
df5
```

```python
### 올림차순이 기본 ascending=True
df1_s = df1.sort_index(ascending=True)
df1_s
```

```python
### 내림차순 하려면 ascending=False
df1_s = df1.sort_index(ascending=False)
df1_s
```

```python
### c컬럼 value값 기준으로 오름차순
df1_c = df1.sort_values(by='c') 
df1_c
```

```python
### c컬럼 value값 기준 내림차순
df1_c = df1.sort_values(by='c',ascending=False) 
df1_c
```

```python
# a 컬럼을 기준으로 올림차순, 내림차순으로 정렬하세요.
df1_a = df1.sort_values(by='a') # 올림차순
df1_a
```

```python
df2_a = df1.sort_values(by='a',ascending=False) # 내림차순
df2_a
```

```python
### 행과 열 바꾸기(전치)
df1_t = df1.transpose()
df1_t
```

```python
# 전치하고 또 전치하면 다시 원위치!
df1_t = df1_t.transpose()
df1_t
```
