```python
# Q. 입력으로 들어오는 모든 수의 평균 값을 계산해주는 함수를 작성한 후 1,2,3을 입력한 결과를 출력하세요.

def avg_number(*args):
    result = 0
    for i in args:
        result += i
    avg = result/len(args)
    return avg

avg_number(1,2,3)
```

name =="main"의미

- 메인 함수의 선언, 시작을 의미합니다. 해당 코드 밑에 main 등의 함수 호출 코드를 작성해서 함수의 기능을 수행
- 직접 mod.py 파일을 실행하는 경우에는 name_ 변수에 main 이라는 값이 할당되고 이 조건절이 참이 되어 if문 다음 문장들이 수행
- 모듈을 불러와서 사용하는 경우 name 변수에는 모듈 이름이 저장되고 이 조건절이 거짓이 되어 if문 다음 문장들이 수행되지 않음
- if name =="main"와 같이 조건문을 사용하여 터미널에서와 같이 직접 호출되어 사용될 때는 그 자체로 기능을 수행하고, 동시에 다른 모듈에서 필요한 함수 등을 제공

```python
with open('mod1.py','w') as f:
    f.write('''
def add(x,y):
    return x + y
    print(add(3,4))
''')
```

```python
with open('mod2.py','w') as f:
    f.write('''
def add(x,y):
    return x + y
if __name__=="main":
    print(add(3,4))
''')
```

```python
# long_task 함수는 수행하는데 5초의 시간이 걸리는 함수이다. 
# 위 프로그램은 이 함수를 수행하는 프로그램이다.
# 이 프로그램은 5초가 5번 반복되니 총 25초의 시간이 걸린다.

import time

def long_task(): # 5초의 시간이 걸리는 함수
    for i in range(5):
        time.sleep(1) # 1초 대기
        print("working:%s\n" %i)
        
print('start')

for i in range(5): # long_task를 5회 수행
    long_task()
    
print('End')
```

```python
import time
import threading  # 여러 일꾼들을 동시에 일 시키기!

def long_task(): # 5초의 시간이 걸리는 함수
    for i in range(5):
        time.sleep(1) # 1초 대기
        print("working:%s\n" %i)
        
print('start')

threads = []
for i in range(5):
    t = threading.Thread(target=long_task)
    threads.append(t)
for t in threads:
    t.start()

for t in threads:
    t.join() # join으로 스레드가 종료할 때까지 기다린다.
    
print('End')
```
