### 적용-결합
데이터 집계(agg),
- 집계 연산을 처리하는 사용자 정의 함수를 그룹 객체에 적용하려면 agg() 메서드 사용
- 모든 열에 여러 함수를 매핑 : group객체.agg([함수1,함수2,함수3,...])
- 각 열마다 다른 함수를 매핑 : group객체.agg({'열1':함수1, '열2':함수2,...})

변환(transform)
- 원소의 본래 행 인덱스와 열 이름을 기준으로 연산 결과를 반환
- 데이터 변환 연산 : group객체.transform(매핑함수)

필터링(filter)
- 그룹 객체에 filter() 메서드를 적용할 때 조건식을 가진 함수를 전달하면 조건이 참인 그룹만을 남김.
- 그룹 객체 필터링 : group 객체.filter(조건함수식)

객체에 함수 매핑(apply)
- 판다스 객체의 개별 원소를 특정 함수에 일대일로 매핑. 사용자가 원하는 대부분의 연산을 그룹 객체에 적용
- group 객체.apply(매핑함수)

```python
# Q. 최대값 - 최소값을 반환하는 사용자 함수를 정의하고 agg()함수에 인수로 전달하여
# 그룹별로 집계하여 agg_minmax 이름으로 저장 후 처음 5개 행을 출력하세요.

import seaborn as sns
titanic = sns.load_dataset('titanic')
df = titanic.loc[:,['age','sex','class','fare','survived']]

# class 열을 기준으로 분할
grouped = df.groupby(['class'])

def min_max(a):
    return a.max()-a.min()

agg_minmax = grouped.agg(min_max) ### <- 하나일 때는 리스트로 안써도 되나봄.
agg_minmax
```

```python
# 모든 열에 여러 함수를 매핑 : group객체.agg([함수1, 함수2, ...])
agg_all = grouped.agg(['min','max'])
agg_all
```

```python
# 각 열마다 다른 함수 매핑 : group객체.agg({'열1':'함수1','열2':'함수2',...})
# ('fare'는 'min','max','age'는 'mean')
agg_all1 = grouped.agg({'age':'mean','fare':['min','max']})
agg_all1
```

```python
# Q. 데이터 개수가 200개 이상인 그룹만을 필터링하여 데이터프레임으로 변환하여
# grouped_filter 이름으로 저장 후 처음 5개 행을 출력하세요.
```

```python
grouped_filter = grouped.filter(lambda x: len(x)>=200)
grouped_filter['class'].value_counts() ### second그룹은 200개가 안 되서 아예 다 빼버린거.
```

```python
# Q. age 열의 평균이 작은 그룹만을 필터링하여 데이터프레임으로 반환하여
# age_filter 이름으로 저장 후 끝에서 5개 행을 출력하세요.

age_filter = grouped.filter(lambda x : x['age'].mean()<30)
age_filter.tail()
```

```python
# z-score : 평균값에서 표준편차의 몇 배 떨어져 있는지를 평가하는 수치
def z_score(x):
    return (x - x.mean()) / x.std()
age_zcore = grouped.age.apply(z_score)
age_zcore.head()
```

```python
age_zscore = grouped.age.transform(z_score)
age_zscore.head()
```

```python
age_filter = grouped.apply(lambda x : x.age.mean()<30)  ### age 평균이 30이하인 그룹들 필터링.

### age_filter 뽑으면 First False / Second True / Third True 나옴.


for x in age_filter.index: ### age_filter.index 는 First, Second, Third
    if age_filter[x] == True:  ### True 는 Second, Third
        age_filter_df = grouped.get_group(x) ### Second / Third 그룹 나눠서
        print(age_filter_df.head())
        print()        
        
# age_filter = grouped.filter(lambda x : x['age'].mean()<30)
# age_filter.tail()
```
